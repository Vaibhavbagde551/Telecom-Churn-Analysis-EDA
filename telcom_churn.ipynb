```python
def convert(value):
    if value == 1:
        return "Yes"
    else:
        return "No"
df['SeniorCitizen'] = df["SeniorCitizen"].apply(convert)
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn': 'count'})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customer")
plt.show()
```

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize = True).unstack()*100

fig, ax  =plt.subplots(figsize=(4,4))

total_counts.plot(kind='bar', stacked = True, ax=ax)

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height /2, f'{height:.1f}%', ha='center', va='center')

plt.title("Churn By Senior Citizen(Stacked Bar Chart)")
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', loc= 'upper right') # customize legend Location
plt.show()
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies']
#number of colums for the subplot grid
n_cols = 3
n_rows = (len(columns) + n_cols - 1)// n_cols  #calulate no. of rows needed)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))

axes = axes.flatten()

#Iterate over colums and plot count plots
for i, col in enumerate(columns):
    sns.countplot(x=col, data = df, ax = axes[i], hue = 'Churn')
    axes[i].set_title(f'Cont Plot of{col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

#Remove empty subplots
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Customers without additional services like Online Backup, Device Protection, Tech Support, and Streaming Services show noticeably higher churn rates.
# Fiber optic users also churn more compared to DSL or no internet services.
# Overall, lack of bundled services is strongly linked with increased churn likelihood.
```
